```python
## Basic Prompt Summarization
from langchain.chat_models import ChatOpenAI
from langchain.schema import(
    AIMessage,
    HumanMessage,
    SystemMessage
)
```

```python
speech="""
Artificial Intelligence (AI) represents the pinnacle of human ingenuity in the realm of technology, embodying the fusion of advanced algorithms, computational power, and the capacity to mimic cognitive functions. At its core, AI aims to enable machines to perceive, reason, learn, and act autonomously, mirroring human intelligence to varying degrees. Its applications span a multitude of domains, from self-driving cars and virtual assistants to medical diagnosis and financial forecasting. As AI continues to evolve and permeate various facets of our lives, it raises profound questions about ethics, privacy, and the future of work. Despite the remarkable progress made in recent years, AI remains a dynamic field ripe with both promise and challenges, shaping the trajectory of technological innovation and human society in profound ways.
"""
```

```python
chat_messages=[
    SystemMessage(content='You are an expert assistant with expertize in summarizing large paragraphs'),
    HumanMessage(content=f'Please provide a short and concise summary of the following paragraph:\n TEXT: {speech}')
]

llm=ChatOpenAI(model_name='gpt-3.5-turbo')
```

```python
##get_summary
print(llm(chat_messages).content)
```

## Text Summarization - Multiple Language

```python

from langchain.chains import LLMChain
from langchain import PromptTemplate
```

```python
generic_template='''
Write a summary of the following speech:
Speech : `{speech}`
Translate the precise summary to {language}.

'''
prompt=PromptTemplate(
    input_variables=['speech','language'],
    template=generic_template
)
```

```python
complete_prompt=prompt.format(speech=speech,language='Hindi')
```

```python
llm_chain=LLMChain(llm=llm,prompt=prompt)
summary=llm_chain.run({'speech':speech,'language':'Hindi'})
```
